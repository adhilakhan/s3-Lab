```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs#used to generate some sample data
from matplotlib import pyplot as plt
```

```python
data,_=make_blobs(n_samples=300) #only need data to be generated.No need of second part
print(data)
```

```python
kmeans=KMeans(n_clusters=5,random_state=42,n_init="auto").fit(data)   #initialization of cluster centroids
labels=kmeans.labels_  #contains the cluster(labels) for each datapoint specifying which cluster each datapoints belongs to
set(labels) #set is a function to set the labels for each datapoint
```

```python
centroids=kmeans.cluster_centers_ #stores the coordinates for 5 cluster
```

```python

plt.scatter(data[:,0],data[:,1],c=labels,cmap="viridis")#this will plot the data points and color in each points according to its cluster labels 
#scatter plot using centroid
plt.scatter(centroids[:,0],centroids[:,1],color="crimson",marker="d")#this will plot the data points and color in each points according to its cluster labels 
plt.legend(list(labels))
plt.show()
```
