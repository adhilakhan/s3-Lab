```python
#position from (1,3),(2,10),(6,10) and finaly(18,20) marker=green,line=red,style=dotted
import numpy as np
import matplotlib.pyplot as plt
x=[1,2,6,18]
y=[3,10,12,20]
plt.title('Line')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='red',linestyle='--',markeredgecolor='green',markerfacecolor='green',marker='o',markersize=10)
plt.show()
```

```python
# axis values from the text file
x=[]
y=[]
with open('axisvalues.txt','r') as file:
    for line in file:
        a,b=map(int,line.strip().split())
        x.append(a)
        y.append(b)
plt.plot(x,y,marker='o')
plt.show()
```

```python
# two or more lines on the same plot with legends
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
z=np.sin(x)
plt.title("Cosine Wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='purple',marker='',markersize='12',label='cos x')
plt.plot(x,z,color='red',marker='',markersize='12',label='sin x')
plt.legend()
plt.show()
```

```python
#multiple plot
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.subplot(3,1,1)
plt.plot(x,y,color='purple',marker='',markersize='12',label='sin x')
plt.title("Sine Wave")
plt.subplot(2,1,2)
plt.plot(x,z,color='purple',marker='',markersize='12',label='cos x')
plt.title("Cosine Wave")
plt.legend()
plt.show()
```

```python
#bar chart
import numpy as np
import matplotlib.pyplot as plt
x=["java","python","php","javascript","c#","c++"]
y=[22.2,17.6,8.8,8,77,6.7]
plt.title('Programming Languages - Popularity')
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.bar(x,y)
plt.show()
```

```python
#horizontal bar chart
plt.barh(x,y,color="red")
```

```python
#horizontal bar chart with different bar color
clr=["violet","indigo","purple","silver","teal","crimson"]
plt.bar(x,y,color=clr)
```
