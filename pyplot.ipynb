```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,11)
y=3*x+5
plt.title("sample Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='orange',marker='o',markersize='12',label='graph')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=[1990,1995,2002,2012,2024]
y=[100,150,9,30,0]
z=[10,26,58,110,126]
plt.title("sample Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='black',marker='o',markersize='12',label='poverty')
plt.plot(x,z,color='blue',marker='o',markersize='12',label='employment')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("Sine Wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='indigo',marker='',markersize='12',label='sin x')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.cos(x)
plt.title("Cosine Wave")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,color='purple',marker='',markersize='12',label='cos x')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
plt.subplot(3,1,1)
plt.plot(x,y,color='purple',marker='',markersize='12',label='sin x')
plt.title("Sine Wave")
plt.subplot(2,1,2)
plt.plot(x,z,color='purple',marker='',markersize='12',label='cos x')
plt.title("Cosine Wave")
plt.legend()
plt.show()
```

```python
x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]
plt.bar(x,y,color='blue',align='center')
plt.bar(x2,y2,color="red",align='center')
plt.title("Bar Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
x=[40,10,20,70,55,65.5]
mylabels = ['Apples','Bananas','Cherries','Dates','Mango','kiwi']
myexplode=[.1,.2,0,0,0,0]
clr=['red','yellow','crimson','blue','orange','green']
plt.pie(x,labels=mylabels,colors=clr,explode=myexplode)
plt.show()
```
